```python
import pandas as pd
from sklearn.decomposition import FactorAnalysis
from sklearn import preprocessing
from pandas.plotting import scatter_matrix
```

```python
factor1 = pd.read_csv("factor1.csv", sep=';', decimal=',', index_col=0)
```

```python
factor1.head()
```

```python
factor1.describe()
```

```python
scatter_matrix(factor1);
```

```python
df_scaled = preprocessing.scale(factor1)
fa = FactorAnalysis(n_components=3)
fa.fit(df_scaled)
```

```python
pd.DataFrame(fa.components_, columns=factor1.columns)
```

```python
print('Uniqueness:')
pd.Series(fa.noise_variance_, factor1.columns)
```

```python
scores = pd.DataFrame(fa.transform(df_scaled), columns=['factor1', 'factor2', 'factor3'])
scores.head()
```

```python
scores.head()
```

```python
factor2 = pd.read_csv('factor2.csv', sep=';', decimal=',')
```

```python
factor2
```

```python
factor2 = factor2.filter(regex='^(?!.*Unnamed).*$')
```

```python
df_scaled = preprocessing.scale(factor2)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=3)
pca.fit(df_scaled)
print(pca.explained_variance_ratio_)
```

```python
print (pca.components_)
```

```python
print ('Explained variance by component: %s' % pca.explained_variance_ratio_)
```

```python
pca.singular_values_
```

```python
(pca.transform(df_scaled))[:5]
```

```python
from sklearn.decomposition import FactorAnalysis

fa = FactorAnalysis(n_components=2)
fa.fit(df_scaled)
```

```python
pd.DataFrame(fa.components_, columns=factor2.columns)
```

```python
print('Uniqueness:')
pd.Series(fa.noise_variance_, factor2.columns)
```

```python
scores = pd.DataFrame(fa.transform(df_scaled), columns=['factor1', 'factor2'])
scores.head()
```
